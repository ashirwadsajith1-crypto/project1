```python
secretnumber=48
lives=5
while lives<=5:
  gno=int(input("Enter the number "))
  diff=abs(secretnumber-gno)
  if diff>=35 and diff<49:
    print("Incorrect!!! Ice Cold" )
  elif diff >=20 and diff<35:
    print("Incorrect!!! Cold")
  elif diff >=10 and diff<20:
    print("Incorrect!!! Warm")
  elif diff >=1 and diff<10:
    print("Incorrect!!! Hot ")
  elif diff==0:
    print("Correct!!! You won")
    print("The secret number is 48")
    break
  else:
    print("...")



```
